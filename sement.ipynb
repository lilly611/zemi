```python
def culc_AIC(df,numlist):
    row_count=len(numlist)
    df_for_X=df.iloc[:,numlist]
    df_for_X.insert(0,"const",1)
    X=df_for_X.to_numpy()
    y_vec=df["y"].to_numpy()
    y_vec=np.array([y_vec]).T
    beta_hat=np.linalg.inv(X.T @ X) @ X.T @ y_vec
    sigma_hat=((y_vec-X@beta_hat).T @ (y_vec-X@beta_hat))/df.shape[0]
    l=-(df.shape[0]*np.log(2*np.pi*sigma_hat)+1)/2
    aic=df.shape[0]*np.log(2*np.pi*sigma_hat)+df.shape[0] + 2*(row_count+2)
    return [numlist,sigma_hat,l,aic]
```

```python
categoli=["組み合わせ","sigma_hat","l","AIC"]
aic_dataframe=pd.DataFrame(index=[],columns=categoli)

num_list=[x for x in range(4)]
for n in range(1,len(num_list)+1):
    for pair in itertools.combinations(num_list,n):
        df2=pd.DataFrame([culc_AIC(df,list(pair))],columns=categoli)
        aic_dataframe=pd.concat([aic_dataframe,df2])
```
